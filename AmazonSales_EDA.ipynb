# Data Cleaning & Preparation

```python
df['actual_price'] = df['actual_price'].str.replace('‚Çπ','')
df['actual_price'] = df['actual_price'].str.replace(',','')
df['actual_price'] = df['actual_price'].astype('float64')

df['discounted_price'] = df['discounted_price'].str.replace("‚Çπ",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = df['discount_percentage'].astype('float64')
```

```python
df['rating'].value_counts()  #Replace '|'
```

```python
df['rating'].mean
```

```python
df['rating'] = df['rating'].str.replace('|','4.2')
df['rating'] = df['rating'].astype('float64')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype('float64')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df[df['rating_count'].isnull()].head()
```

```python
df['rating_count'] = df.rating_count.fillna(value=df['rating_count'].median())
```

# Data Visualization

```python
df['rating'].plot(kind = 'hist', color = 'orange', edgecolor = 'black', bins = 10)
plt.xlabel('Rate')
plt.show()
```

```python
df_categ = df['category'].str.split('|').str[0]
for i in range(len(df_categ)):
    if df_categ[i] not in ['Electronics', 'Computers&Accessories', 'Home&Kitchen']:
        df_categ[i] = 'Others'
```

```python
df_categ.value_counts().plot(kind = 'pie', explode = [0.02,0.03,0.02,0.1] ,autopct = '%1.1f%%')
plt.title('Categories')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
top_rated = df[df['rating_count'] >= 100].sort_values(by='rating', ascending=False).dropna().head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='rating', y='product_name', data=top_rated, palette="crest")
plt.title('Top 10 Rated Products (min. 100 reviews)')
plt.xlabel('Rating')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['discount_percentage'], bins=30, color='green', kde=True)
plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

```python
price_categ = []
for p in df['actual_price']:
    if p <= 3000:
        price_categ.append('Low Budget Price')
    elif p <= 10000:
        price_categ.append('Mid Range Price')
    else:
        price_categ.append('High End Price')

price_disc_categ = []
for p in df['discounted_price']:
    if p <= 3000:
        price_disc_categ.append('Low Budget Price')
    elif p <= 10000:
        price_disc_categ.append('Mid Range Price')
    else:
        price_disc_categ.append('High End Price')

# Define the desired order of categories
category_order = ['Low Budget Price', 'Mid Range Price', 'High End Price']

# Compute the counts for each category
actual_counts = pd.Series(price_categ).value_counts().reindex(category_order)
discounted_counts = pd.Series(price_disc_categ).value_counts().reindex(category_order)

# Create the subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Plot the bar chart for Actual Prices
axs[0].bar(range(len(category_order)), actual_counts, edgecolor='black')
axs[0].set_title("Actual Prices")
axs[0].set_xlabel("")
axs[0].set_ylabel("Frequency")
axs[0].set_xticks(range(len(category_order)))
axs[0].set_xticklabels(category_order)

# Plot the bar chart for Discounted Prices
axs[1].bar(range(len(category_order)), discounted_counts, color='red', edgecolor='black')
axs[1].set_title("Discounted Prices")
axs[1].set_xlabel("")
axs[1].set_ylabel("Frequency")
axs[1].set_xticks(range(len(category_order)))
axs[1].set_xticklabels(category_order)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize = (14,5))
axs[0].scatter(df['actual_price'], df['rating'])
axs[0].set_title("")
axs[0].set_xlabel("Actual Prices")
axs[0].set_ylabel("Rate")

axs[1].scatter(df['discounted_price'],df['rating'], color = 'red')
axs[1].set_title("")
axs[1].set_xlabel("Discounted Price")
axs[1].set_ylabel("Rate")

plt.tight_layout()
```
